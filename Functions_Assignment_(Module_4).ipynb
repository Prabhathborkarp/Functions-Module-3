####Q1. What is the difference between a function and a method in Python?

####Answer:

A function in Python is a block of code that helps us do a specific task. We give it some input, and it gives us back a result. Functions work on their own and don’t need to be inside a class. For example:

    def say_hello(name):
    return f"Hello, {name}!"

We can call this function like this:

    print(say_hello("Alice"))
    This will print: Hello, Alice!

A method is also a function, but it lives inside a class and works with objects. It can use or change data that belongs to the object. Here's an example:

    class Greeter:
    def say_hello(self, name):
    return f"Hello, {name}!"

To use this method, we first create an object from the class:

    g = Greeter()
    print(g.say_hello("Alice"))

This also prints: Hello, Alice!, but now we’re using a method that belongs to the object g.

So, the difference is:

  - A function is used on its own.

  - A method is used with an object from a class.

Both help us organize our code and reuse it more easily.


####Q2. Explain the concept of function arguments and parameters in Python.

####Answer:

In Python, when we create a function, we can pass data into it so the function can use that data. The terms parameters and arguments help describe how that data is handled.

Parameters are the names we use in the function definition to receive data. They are like empty containers that wait to be filled when we call the function.

For example:

    def greet(name):
    print(f"Hello, {name}!")

Here, name is a parameter of the function greet.

Arguments are the actual values we give to the function when we call it. They fill the parameters with real data.

Example:
    
    greet("Alice")

Here, "Alice" is an argument that is passed into the parameter name.

    def add(a, b):  # a and b are parameters
    return a + b

We pass arguments when we call the function:

    result = add(3, 5)  # 3 and 5 are arguments
    print(result)       # Output: 8

In simple terms:
Parameters = what the function expects

Arguments = what we actually give it

####Q3. What are the different ways to define and call a function in Python?

####Answer:

In Python, we can define and call functions in several different ways depending on what we want the function to do and how flexible we want it to be. Here’s a simple explanation of the different ways to define and call functions:

1. Normal Function (with required parameters)

  - Define:

          def greet(name):
          print(f"Hello, {name}!")
  - Call:

          greet("Alice")
The function requires one argument when we call it.

2. Function with Default Parameters

  - Define:

          def greet(name="friend"):
          print(f"Hello, {name}!")
  - Call:

          greet()  # Output: Hello, friend!
          greet("Bob")      # Output: Hello, Bob!

    If we don’t pass an argument, the default value is used.


3. Function with Multiple Parameters
  - Define:

          def add(a, b):
          return a + b
  - Call:

          print(add(3, 5))     # Output: 8
    We must pass all required arguments.

4. Function with Keyword Arguments
  - Define:

          def introduce(name, age):
          print(f"My name is {name} and I'm {age} years old.")
  - Call:

          introduce(age=25, name="Alice")

    We can specify which value goes to which parameter using keywords.

5. Function with Variable Number of Arguments (*args)
  - Define:

          def add_all(*numbers):
          return sum(numbers)
  - Call:

          print(add_all(1, 2, 3))       # Output: 6
          print(add_all(5, 10, 15, 20)) # Output: 50
We can pass any number of values. They’re received as a tuple.

6. Function with Keyword Variable Arguments (**kwargs)
  - Define:

          def print_info(**info):
          for key, value in info.items():
          print(f"{key}: {value}")
  - Call:

          print_info(name="Alice", age=30)

    We can pass any number of keyword arguments. They’re received as a dictionary.

7. Lambda (Anonymous) Functions

  - Define and call:

          square = lambda x: x * x
          print(square(4))   # Output: 16

    Useful for short, simple functions.

#### Q4. What is the purpose of the `return` statement in a Python function?

When we use return, the function finishes its job and gives us a value that we can store or use elsewhere in our program. For example:

    def add(a, b):

Here, the function calculates the sum of a and b, and the return statement gives us that result. We can then use it like this:

    result = add(3, 5)

In this case, add(3, 5) returns the value 8, which we store in result. If we didn’t have the return statement, the function would do the calculation, but we wouldn't be able to capture the result or use it in other parts of our program. So, in simple terms, the return statement allows a function to output a value that we can use elsewhere.

####Q5. What are iterators in Python and how do they differ from iterables?

#### Answer:
In Python, iterators and iterables are closely related but serve different purposes. Let’s understand them in a simple way:

An iterable is any object that can be looped over using a for loop. Examples of iterables include lists, strings, tuples, and dictionaries. These objects contain data that we can go through one item at a time. However, an iterable by itself does not remember where it is in the sequence — it just provides access to the data.

An iterator is an object that remembers its position during iteration and gives us the next item when we ask for it. We get an iterator by using the iter() function on an iterable. Once we have an iterator, we can get the next item using the next() function. The iterator continues until there are no more items, and then it raises a StopIteration error.

For example:

    numbers = [1, 2, 3]  # This is an iterable
    it = iter(numbers)   # This is an iterator

    print(next(it))  # Output: 1
    print(next(it))  # Output: 2
    print(next(it))  # Output: 3

If we call next(it) again, it will raise a StopIteration error because there are no more items.

Key Differences:

  - An iterable can be looped over (like a list or string), but it doesn’t keep track of where it is.

  - An iterator is the tool that actually does the looping — it keeps track of the current position and knows how to get the next item.

In short, an iterable gives us data, and an iterator helps us go through that data one step at a time.

#### Q6. Explain the concept of generators in Python and how they are defined.

####Answer:
Generators in Python are a special kind of function that allow us to create iterators in a simple and memory-efficient way. Instead of computing and storing all the values at once (like a list), generators produce values one at a time only when needed, which is great for working with large data or infinite sequences.

We define a generator just like a normal function, but instead of using return to send back a result and finish, we use the yield keyword. Each time the generator’s yield statement runs, it outputs a value and pauses its state, so when we ask for the next value, it continues right where it left off.

For example:

    def count_up_to(n):
    count = 1
    while count <= n:
    yield count
    count += 1

When we call count_up_to(5), it returns a generator object. We can then iterate over it:


    for number in count_up_to(5):
    print(number)

This prints numbers from 1 to 5, generating each number only when the loop asks for it, rather than creating the whole list at once.

In summary, generators let us write functions that produce a sequence of values lazily, saving memory and allowing efficient iteration, especially when dealing with large or infinite data. They are defined using the yield keyword instead of return.

####Q7. What are the advantages of using generators over regular functions?

####Answer:
Generators offer several advantages over regular functions, especially when dealing with sequences of data:

1. Memory Efficiency: Unlike regular functions that return the entire result at once (like a list), generators produce one item at a time and only when needed. This means they use much less memory, which is particularly helpful when working with large datasets or infinite sequences.

2. Lazy Evaluation: Generators generate values on the fly, pausing after each yield and resuming only when the next value is requested. This lazy evaluation avoids unnecessary computation and can improve performance when we don’t need all the results immediately.

3. Improved Performance: Because generators produce values one at a time, they often start yielding results faster than a function that builds and returns a complete data structure, enabling smoother processing in data pipelines or streaming data scenarios.

4. Simpler Code for Iteration: Generators provide a clean and readable way to create iterators without having to write a full class with __iter__() and __next__() methods, reducing boilerplate code.

5. Ability to Represent Infinite Sequences: Generators can represent sequences without a defined end because they produce values as requested, which is impossible with regular functions returning fixed-size data.

Regular function (returns a list):

    def get_squares(n):
    result = []
    for i in range(n):
    result.append(i * i)
    return result

    squares = get_squares(5)
    print(squares)  # Output: [0, 1, 4, 9, 16]

This function creates a list of squares from 0 to 4 all at once and returns it. If n is very large, this list could take a lot of memory.

Generator function (yields values one by one):

    def generate_squares(n):
    for i in range(n):
    yield i * i

    squares_gen = generate_squares(5)
    for square in squares_gen:
    print(square)

    Output:

    0
    1
    4
    9
    16
    
In this case, the generator produces each square one at a time only when the loop requests it. This way, memory is saved because the whole list is never stored at once. If n were huge or infinite, the generator would still work efficiently, whereas the regular function might use too much memory or crash.

So, generators let us process data step-by-step, saving memory and allowing better performance for large or complex tasks.

####Q8. What is a lambda function in Python and when is it typically used?

####Answer:

A lambda function in Python is a small, anonymous function defined with the lambda keyword instead of the usual def. It can have any number of input parameters but only one expression, which is evaluated and returned automatically. Lambda functions are often used for simple, quick tasks where defining a full function with def feels unnecessary.

For example:

    square = lambda x: x * x
    print(square(5))  # Output: 25

Here, lambda x: x * x creates a function that squares its input, and we assign it to the variable square.

Lambda functions are typically used when we need a small function for a short time, especially as an argument to other functions like map(), filter(), or sorted(). They help write concise code without cluttering the program with many named functions.

For instance, sorting a list of tuples by the second element:

    pairs = [(1, 3), (2, 1), (3, 2)]
    pairs.sort(key=lambda x: x[1])
    print(pairs)  # Output: [(2, 1), (3, 2), (1, 3)]

In summary, lambda functions are handy for creating simple, unnamed functions on the fly, usually for short-term use in other function calls or when we want concise code.

####Q9. Explain the purpose and usage of the `map()` function in Python.

####Answer:

The map() function in Python is used to apply a given function to every item of an iterable (like a list, tuple, etc.) and return a new iterator with the results. This allows us to transform or process all elements in a collection efficiently without writing an explicit loop.

The basic usage of map() looks like this:

    map(function, iterable)
  - function is the function we want to apply to each item. It can be a regular function or a lambda function.

  - iterable is the collection of items we want to process.

For example, if we want to square every number in a list, we can do:

    numbers = [1, 2, 3, 4, 5]
    squares = map(lambda x: x * x, numbers)
    print(list(squares))  # Output: [1, 4, 9, 16, 25]

Here, map() applies the lambda function x * x to each number in the list and returns an iterator of squares. We convert it to a list to see all results at once.

The advantage of map() is that it makes the code cleaner and more readable by removing the need for explicit loops when applying the same operation to many items. It is also memory efficient because it returns a generator-like object that produces results on demand, rather than creating a whole new list immediately.

In short, map() helps us transform data collections quickly and elegantly by applying a function to each element.

####Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

####Answer:

1. map()
Purpose: Applies a given function to each item in an iterable and returns an iterator with the results.

  Usage: Transform or modify every element in a collection.

  Example:

        numbers = [1, 2, 3, 4]
        squares = map(lambda x: x*x, numbers)
        print(list(squares))  # Output: [1, 4, 9, 16]

2. filter()
Purpose: Applies a function that returns True or False to each item in an iterable, and returns only the items where the function returns True.

  Usage: Select or filter elements based on a condition.

  Example:

        numbers = [1, 2, 3, 4, 5]
        evens = filter(lambda x: x % 2 == 0, numbers)
        print(list(evens))  # Output: [2, 4]

3. reduce() (from functools module)
Purpose: Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value.

  Usage: Combine all elements into one result by repeatedly applying a function.

  Example:

    from functools import reduce

        numbers = [1, 2, 3, 4]
        sum_all = reduce(lambda x, y: x + y, numbers)
        print(sum_all)  # Output: 10

  
Use map() when we want to transform all items.

Use filter() when we want to keep only some items.

Use reduce() when we want to combine all items into one.

####Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

####Answer:

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example usage
my_list = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(my_list)
print("Sum of even numbers:", result)
```

####Q4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

####Answer:

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def check_primes_1_to_200():
    for number in range(1, 201):
        if is_prime(number):
            print(f"{number} is a prime number")
        else:
            print(f"{number} is not a prime number")

# Run the function
check_primes_1_to_200()
```

####Q5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

####Answer:

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Total number of terms to generate
        self.index = 0      # Current index
        self.a = 0          # First Fibonacci number
        self.b = 1          # Second Fibonacci number

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.terms:
            raise StopIteration
        if self.index == 0:
            self.index += 1
            return self.a
        elif self.index == 1:
            self.index += 1
            return self.b
        else:
            next_val = self.a + self.b
            self.a, self.b = self.b, next_val
            self.index += 1
            return next_val

# Ask user for input
num_terms = int(input("Enter the number of Fibonacci terms to generate: "))

# Create iterator and print Fibonacci sequence
fib_iter = FibonacciIterator(num_terms)
for num in fib_iter:
    print(num)
```

####Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

####Answer:

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Ask user to enter the exponent value
max_exp = int(input("Please enter the exponent value (non-negative integer): "))

# Generate and print powers of 2 up to the provided exponent
for value in powers_of_two(max_exp):
    print(value)
```

####Q7.  Implement a generator function that reads a file line by line and yields each line as a string.

####Answer:

```python
def file_line_generator(filepath):
    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Ask the user for the filename
filename = input("Please enter the filename to read: ")

# Use the generator to print each line
for line in file_line_generator(filename):
    print(line, end='')
```

####Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

####Answer:

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40]

# Use map() to convert each Celsius temperature to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Print the result
print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

####Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

####Answer:

```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    filtered_chars = filter(lambda char: char not in vowels, s)
    return ''.join(filtered_chars)

# Ask user for input string
input_string = input("Enter a string: ")

# Remove vowels and print result
result = remove_vowels(input_string)
print("String without vowels:", result)
```
